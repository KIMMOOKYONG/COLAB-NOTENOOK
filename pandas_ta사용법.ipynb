```python
import pandas as pd
import pandas_ta as ta
import yfinance as yf
import talib
import warnings
warnings.filterwarnings("ignore")
```

# 데이터 다운로드

```python
# 주가 데이터 다운로드 및 이동평균 칼럼 초기화
data = yf.download("AAPL", start="2020-01-01")
data.head()
```

# 단순이동평균계산

## pandas_ta 직접 참조 방식

```python
# pandas 프레임 사용하지 않는 경우, 임의 칼럼에 대해서 이동평균 계산이 가능하다.
# 계산할 칼럼을 임의로 설정 가능
sma = ta.sma(close=data["Close"], length=2, offset=None)
data.head()
```

## dataframe pandas-ta 간접 참조 방법

```python
# PANDAS-TA 라이브러리를 적용하려면, 반드시 데이터 칼럼구조가 OHLC 구조를 가져야 한다.
# 모든 계산에 대해서 참조하는 값은 Close 칼럼이다.
data.ta.sma(length=20, offset=None, append=True)
data.dropna().head()
```
