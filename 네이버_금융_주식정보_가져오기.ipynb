# 종목 목록 수집

```python
today = datetime.today().strftime("%Y%m%d")
```

```python
ticker_list = stock.get_market_ticker_list(date = today, market="KOSPI")
ticker_list
```

```python
tickers = None
symbols = None
symbol_list = []
```

```python
for ticker in ticker_list:
    symbol = stock.get_market_ticker_name(ticker)
    symbol_list.append(symbol)
```

```python
len(ticker_list), len(symbol_list)
```

```python
# How to Add / Insert a Row into a Pandas DataFrame
# https://datagy.io/pandas-add-row/
```

```python
# 주식 거래가 이루어지지 않는 날의 데이터를 요청하면 반환되는 정보가 없다.
data = pd.DataFrame(columns=["BPS", "PER","PBR","EPS","DIV","DPS"])
from_date = "20220701"
for ticker in tqdm(ticker_list):
    data = data.append(stock.get_market_fundamental_by_date(from_date, today, ticker))
```

```python
data["회사명"] = symbol_list[:len(ticker_list)]
```

```python
data = data[["회사명", "BPS", "PER", "PBR", "EPS", "DIV", "DPS"]]
data
```
